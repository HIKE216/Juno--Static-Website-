```python
oldPrice = []
price = soup('div', class_ = 'old')

for i in price:
    oldPrice.append(i.text)

oldPrice
```

```python
discountPrice = []
price = soup('div', class_ = 'bdg _dsct _sm')

for i in price:
    discountPrice.append(i.text)

discountPrice
```

```python
# for all the pages
pages = [x for x in range(1,51)]

currentPrice = []
oldPrice = []
discountPrice = []
laptopnames = []

for i in pages:
    url = f"https://www.jumia.com.ng/catalog/?q=laptops&page={i}#catalog-listing"
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "lxml")

# ProductName
    names = soup.find_all('h3', class_ = 'name')
    for j in names:
        laptopnames.append(j.text)

    # CurrentPrice
    cprice = soup('div', class_ = 'prc')
    for k in cprice:
        currentPrice.append(k.text)

    # Discount
    dprice = soup('div', class_ = 'bdg _dsct _sm')
    for l in dprice:
        discountPrice.append(l.text)

    # Old Price
    oprice = soup('div', class_ = 'old')
    for m in oprice:
        oldPrice.append(m.text)


data = {'title': laptopnames, 'currentPrice':currentPrice, 'oldPrice': oldPrice,'discount': discountPrice}

laptop_df = pd.DataFrame.from_dict(data,orient= 'index').transpose()
laptop_df
```

### Data Transformation and Loading

```python
laptop_df.info()
```

```python
laptop_df.fillna({
    'oldPrice': '0',
    'discount' : '0%'
}, inplace=True)
```

```python
laptop_df.info()
```

```python
# define database parameters
db_params = {
    'username': 'postgres',
    'password': '******',
    'host': 'localhost',
    'port': '5432',
    'database': 'juno_ecommerce'
}

db_url = f"postgresql://{db_params['username']}:{db_params['password']}@{db_params['host']}:{db_params['port']}/{db_params['database']}"

engine = create_engine(db_url)

# define to PostgreSQL server
with engine.connect() as connection:
    # Create table and load the data
    laptop_df.to_sql('junoecommerce_products',connection,index = False, if_exists='replace')

print('Database created successfully')
```
